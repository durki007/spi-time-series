# Data Loading &amp; Initial Inspection — BPI Challenge 2017

This notebook downloads and loads the **BPI Challenge 2017** event log using the
`Dataset` class from the `spi_time_series` package. It provides the first
point of contact with the raw data before any preprocessing or modelling.

**What you'll find here:**
- Automatic download &amp; caching of the XES log via `Dataset`
- Quick glance at the log structure (`head`, `describe`)
- Verification that the data is ready for the pipeline

## 1. Imports &amp; Logging

We import the `Dataset` class from `spi_time_series` and configure basic logging
so we can see download progress and any warnings.

```python
import logging

from spi_time_series.data import Dataset

logging.basicConfig(level=logging.INFO)
```

## 2. Load the Dataset

`Dataset()` automatically downloads the BPI Challenge 2017 XES log (if not already
cached) and exposes it as a `pandas.DataFrame` via the `.log` property.

```python
ds = Dataset()
ds.log.head()
```

## 3. Quick Inspection

A `describe(include="all")` gives us column types, cardinality, and basic
statistics to verify the log was loaded correctly.

```python
ds.log.describe(include="all")
```
